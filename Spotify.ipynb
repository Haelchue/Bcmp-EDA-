```python
# 6. Genre Clustering
print("\n6. GENRE CLUSTERING\n")

# Import required libraries
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Prepare data for clustering
cluster_features = ['danceability_%', 'energy_%', 'valence_%', 'acousticness_%']
X = df[cluster_features].dropna()

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Elbow method to find optimal clusters
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(10,6))
plt.plot(range(1,11), wcss, marker='o')
plt.title('Elbow Method for Optimal Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()

# Perform clustering (assuming 4 clusters based on elbow)
kmeans = KMeans(n_clusters=4, random_state=42)
clusters = kmeans.fit_predict(X_scaled)

# Add clusters back to original dataframe (only for rows with no missing values)
valid_indices = X.index
df.loc[valid_indices, 'cluster'] = clusters

# Analyze clusters
cluster_analysis = df.groupby('cluster')[cluster_features].mean()
print("Cluster Characteristics:")
display(cluster_analysis)

# Visualize clusters
plt.figure(figsize=(10,6))
sns.scatterplot(data=df.dropna(subset=['cluster']), 
                x='danceability_%', 
                y='energy_%', 
                hue='cluster', 
                palette='viridis')
plt.title('Song Clusters by Audio Features')
plt.show()

# Additional visualization with more features
plt.figure(figsize=(12,8))
sns.pairplot(data=df.dropna(subset=['cluster']), 
             vars=cluster_features, 
             hue='cluster', 
             palette='viridis',
             plot_kws={'alpha': 0.6})
plt.suptitle('Pairwise Feature Relationships by Cluster', y=1.02)
plt.show()
```
